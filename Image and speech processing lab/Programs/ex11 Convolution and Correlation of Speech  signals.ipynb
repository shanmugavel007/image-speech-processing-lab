## Ex11 Convolution and Correlation of Speech  signals

### a) Convolution of Speech signals

```python
from scipy import signal
import numpy as np 
import librosa 
import matplotlib.pyplot as plt 

d_file = "./auds/set1.wav"
d1_file="./auds/set2.wav"

d, sr = librosa.load(d_file,8000)
d1, sr = librosa.load(d1_file,8000)
plt.subplot(2,2,1) 
t=np.linspace(0,len(d)/sr,len(d))
plt.plot(t,d, lw=1)
y=signal.convolve(d,d1)
plt.grid(True)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.subplot(2,2,2) 
t=np.linspace(0,len(d1)/sr,len(d1)) 
plt.plot(t,d1, lw=1)
plt.grid(True)
plt.xlabel('Time(s)')
plt.ylabel(' Amplitude')
plt.subplot(2,2,3)
t=np.linspace(0,len(y)/sr,len(y)) 
plt.plot(t,y, lw=1)
plt.grid(True) #Naming the x-axis, y-axis 
plt.title('Covolution of signals') 
plt.xlabel('Time(s)') 
plt.ylabel(' Amplitude') #Adds space between different subplots 
plt.tight_layout()
plt.grid(True) #To load the display window 
plt.show()
```

### b)  Corelation of Speech signals

```python
#b
# Importing the libraries.
import numpy as np
import librosa
import matplotlib.pyplot as plt
d_file = "./auds/speech.wav"
d, sr = librosa.load(d_file,8000)
plt.subplot(2,1,1)
t=np.linspace(0,len(d)/sr,len(d))
plt.plot(t,d, lw=1)
plt.xlabel('Time in seconds')
plt.ylabel('Amplitude')
#computing autocorrelation using librosa
r = librosa.autocorrelate(d, max_size=10000)
plt.subplot(2,1,2)
print(r.shape)
# Adding plot title
plt.title("Autocorrelation Plot")
# Providing x-axis name.
plt.xlabel("Lags")
plt.ylabel('Coefficients')
# Plotting the Autocorreleation plot
plt.acorr(r, maxlags = 20)
# Displaying the plot
print("The Autocorrelation plot for the data is:")
plt.grid(True)
plt.tight_layout(pad=3.0)
#plt.tight_layout()
plt.show()
```

### Cross Correlation

```python
#crosscorrelation
import librosa
import matplotlib.pyplot as plt
from scipy import signal
import numpy as np
d_file = "./auds/sample1.wav"
d1_file="./auds/sample2.wav"
#load audio files with librosa
d, sr = librosa.load(d_file,8000)
d1, sr = librosa.load(d1_file,8000)
#t=np.linspace(0,len(d)/sr,len(d))
#plt.plot(t,d, lw=1)
#print array values of first .wav file
print(d)
#print array values of second .wav file
print(d1)

correlation = np.correlate(d, d1, mode="full")
lags = np.arange(-100, 100) 
print(correlation) 
plt.subplot(2,2,1)
plt.title('speech file1') 
t=np.linspace(0,len(d)/sr,len(d))
plt.plot(t,d, lw=1)
plt.grid(True) 
plt.xlabel('Time(s)')
plt.ylabel('Amplitude') 
plt.subplot(2,2,2)
plt.title('speech file2')
t=np.linspace(0,len(d1)/sr,len(d1)) 
plt.plot(t,d1, lw=1)  
plt.grid(True) 
plt.xlabel('Time(s)')
plt.ylabel('Amplitude') 
plt.subplot(2,2,3)
plt.plot(lags,correlation[0:200]) 
plt.title("crosscorrelation Plot") 
plt.xlabel("Lags") 
plt.ylabel('Coefficients') 
plt.grid(True) 
plt.tight_layout() 
plt.show()
```
