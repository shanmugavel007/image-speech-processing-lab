## EX:1 smoothening and sharpening of image

### a)smoothening of image

```python
#a
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('./imgs/lady.jpg')
blur = cv2.blur(img,(5,5))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

### b)sharpening of image

```python
#sharpening
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('./imgs/lady2.jpg',1)
# Creating our sharpening filter
filter = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
# Applying cv2.filter2D function on our Cybertruck image
sharpen_img=cv2.filter2D(img,-1,filter)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(sharpen_img),plt.title('Shapening Image')
plt.xticks([]), plt.yticks([])
plt.show()
```
