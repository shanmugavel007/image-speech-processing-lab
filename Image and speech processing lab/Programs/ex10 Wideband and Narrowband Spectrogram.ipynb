## Ex10 Wideband and Narrowband Spectrogram

```python
import matplotlib.pyplot as plt #Drawing
import librosa.core as lc #Calculate the use of stft
import numpy as np #Used some of the tool functions
import librosa.display #Picture sound spectrogram
```

```python
path = "./auds/speech.wav"
y, sr = librosa.load(path,8000)
n_fft = 1024 #FFTLength
#Plotting original speech signal
plt.subplot(2,1,1)
plt.title('Original Speech Signal')
plt.plot(y,label="original file") # actual ploting
plt.xlabel("Time") # label of x-axis
plt.grid(True)
#And plot the results
fig, ax = plt.subplots()
#Get wideband spectrogram
mag = np.abs(lc.stft(y,n_fft=n_fft,hop_length=10,win_length=40,window='hamming')) 
mag1=librosa.amplitude_to_db(mag, ref=np.max) #mapping the magnitudes to a decibel scale.
img=librosa.display.specshow(mag1,sr=sr,x_axis='s', y_axis='linear')#display spectrogram
ax.set(title='Wideband spectrogram')
fig.colorbar(img, ax=ax, format="%+2.f dB")
#Get narrowband spectrogram
fig, ax = plt.subplots()
mag1 = np.abs(lc.stft(y,n_fft=n_fft, hop_length=100, win_length=400, window='hamming'))
#STFT of y
D1 = librosa.amplitude_to_db(mag1, ref=np.max)#mapping the magnitudes to a decibel scale.
img1=librosa.display.specshow(D1, sr=sr,x_axis='s', y_axis='linear')#display spectrogram
ax.set(title='narrowband spectrogram')
fig.colorbar(img1, ax=ax, format="%+2.f dB")
plt.tight_layout()
plt.show()
```
